# Project Assignment 2

```python
# Verify Missing Values
project_df.isnull().sum()
```

```python
# Eliminate observations missing key data
project_df = project_df.dropna(subset=["Yds", "Att", "TD", "Int", "Rate"])
```

```python
# Create minimum games played/started to count as a "Startin QB"
starter_df = project_df[
 (project_df["G"] >= 8) &
 (project_df["GS"] >= 6) &
 ((project_df["Yds"] >= 1500) | (project_df["Att"] >= 200))
].copy()
```

```python
starter_df["Era"] = pd.cut(
 starter_df["Year"],
 bins=[2000, 2009, 2016, 2023],
 labels=["Early 2000s", "2010–2016", "Modern Era"])
```

```python
# Final Clean Dataset
starter_df.to_csv("starter_qb_cleaned.csv", index=False)
```

```python
# Clean player names
starter_df["Player_clean"] = (
    starter_df["Player"]
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)
```

```python
# League Avg Each Season
avg_by_year = (
    starter_df
    .groupby("Year")[["Yds","Rate","TD","Int"]]
    .mean()
    .reset_index()
)

avg_by_year.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(avg_by_year["Year"], avg_by_year["Yds"], marker="o")

plt.title("Average Passing Yards per Season")
plt.xlabel("Year")
plt.ylabel("Average Passing Yards")

plt.show()
```
